Q3.

Steps to create average order value

**1.** open Power BI and select dataset from the field pane.

**2.** click on new measure.

table tools > New measure.

**3.** Enter the DAX formula in the formula bar.

       Average Order Value = DIVIDE(SUM(Table1[SalesAmount]), DISTINCTCOUNT(Table1[OrderID]))

**4.** verify the measure

- The measure Average Order Value will appear in the Fields pane with a calculator icon, indicating that it is a measure.

**5.** Display the Measure

- Insert a Card Visual from the Visualizations pane.

Q4.

steps to create a measure that calculates cumulative Year-to-Date (YTD) Sales based on the Order Date.

**1.** Open Power BI and go to:

- Report View or Data View

**2.** Select the dataset table from the field pane.

**3.** click "new measure" on the ribbon

"table tools > new measure"

**4.** Enter the DAX Formula

       YTD Sales = TOTALYTD(SUM([SalesAmount]), 'TableName'[OrderDate])

press "ENTER"

*After creating the measure, it will appear in the Fields pane with a calculator icon. Since a measure is a dynamic calculation, it cannot be viewed as a column in Data View. To see the results, add the YTD Sales measure to a Table, Matrix, or Line Chart visual along with the OrderDate field.*

Q5.

steps to create product rank based on total sales in descending order.

**1.** open the dataset in power BI.

**2.** From the table view option create a measure of total sales.

- Table Tools > New Measure

**3.** Enter the DAX formula in the formula bar

       Total Sales = SUM(Table1[SalesAmount])

press "ENTER"       

**4.** Create a New Measure for Ranking

Again click:

Table Tools > New Measure

**5.** Enter the DAX Formula

     Product Rank =
     RANKX(
        ALL(Table1[ProductID]),
        [Total Sales],
        ,
        DESC,
        DENSE
    )

Press "ENTER"

**6.** Verify the Measure

- The measure will appear with a calculator icon in the Fields pane.

**7.** Display the Ranking

- Insert a Table Visual.

- Add:
  
  ProductID

  Total Sales
  
  Product Rank

- Sort the table by Product Rank ascending.

Q6.

Steps to create a Calculated Column Classifying Orders as "High Value" if Sales > 2000, Else "Low Value"

**1.** Open table view tab in Power BI

**2.** Select the dataset (Amazon Store Sales Dataset)

**3.** Click new column.

Table Tools > New Column

**4.** Enter the DAX Formula

       Order Value Category =
       IF(
           [SalesAmount] > 2000,
           "High Value",
           "Low Value"
       )

Press "ENTER"

**5.** Verify the New Column

Q7.

Steps to create a Measure to Calculate Total Sales from the Online Channel Only.

**1.** in power bi table view tab open the report containing the Amazon Store Sales dataset.

**2.** Select the data set (Amazon Store Sales Dataset)

**3.** Click new measure.

Table Tools > New Measure

**4.** Enter the DAX Formula.

       Online Sales =
       CALCULATE(
            SUM(Table1[SalesAmount]),
            Table1[SalesChannel] = "Online"
       )

Press "ENTER"
